## Histograma Simples e Comparativo

```python
# Esta funo gera um histograma simples para um conjunto de dados unidimensional.
# Parmetros:
# - data: O conjunto de dados para o qual queremos criar o histograma.
def HistogramaSimples(data):
    # Cria um histograma com 20 bins, com as barras em azul e bordas pretas.
    plt.hist(data, bins=20, color='blue', edgecolor='black')
    plt.title('Distribuio de Dados')  # Adiciona ttulo ao grfico.
    plt.xlabel('Valores')  # Adiciona rtulo ao eixo x.
    plt.ylabel('Frequncia')  # Adiciona rtulo ao eixo y.
    plt.grid(True)  # Adiciona grade ao grfico.
    plt.show()  # Exibe o grfico.

# Esta funo gera um histograma comparativo para mltiplos conjuntos de dados.
# Parmetros:
# - data: Uma lista de arrays (cada array representa um conjunto de dados).
def HistogramaComparativo(data):
    # Define o tamanho da figura do grfico.
    plt.figure(figsize=(10,5))
    plt.title('Distribuio de Dados')  # Adiciona ttulo ao grfico.
    plt.xlabel('Valores')  # Adiciona rtulo ao eixo x.
    plt.ylabel('Frequncia')  # Adiciona rtulo ao eixo y.
    plt.grid(True)  # Adiciona grade ao grfico.

    # Itera sobre os arrays de dados na lista 'data'.
    for i in range(len(data)):
        # Cria um histograma para cada conjunto de dados, com as barras em azul,
        # bordas pretas e com transparncia de 0.5 para sobreposio visual.
        plt.hist(data[i], bins=20, color='red', edgecolor='black', alpha=0.5)
```

##  KOLMOGOROV-SMIRNOV

```python
def KolmogorovSmirnov(rol, alpha, tamanho, graf=True):
    """
    Funo para realizar o teste de Kolmogorov-Smirnov.

    Parmetros:
        rol (array): Array de dados.
        alpha (float): Nvel de significncia.
        tamanho (int): Tamanho da amostra.
        graf (bool, opcional): Indica se deve exibir grficos e informaes. Padro  True.

    Retorna:
        bool: True se a hiptese nula  aceita, False caso contrrio.
    """

    print('\nKolmogorov-Smirnov\n')

    # H0 = Fobs(x) = Fesp(x)
    # H1 = Fobs(x) != Fesp(x)

    # Dcalc = max{|Fesp(x) - Fob(x)|,|fsp(x) - Fob(x-1)|}
    # Fesp = Frequencia relativa
    # Fob = Frequencia relativa acumulada
    Media = rol.mean()
    Variancia = rol.var()
    DesvioPadrao = rol.std()
    mediana = np.median(rol)
    min = rol.min()
    maximum = rol.max()
    soma = rol.sum()

    if graf:
        HistogramaSimples(rol)
        print('Media = ', Media)
        print('Variancia = ', Variancia)
        print('DesvioPadrao = ', DesvioPadrao)
        print("Mediana = ", mediana)
        print("Minimo = ", min)
        print("Maximo = ", maximum)
        print("Soma = ", soma)
        print("Contagem = ", tamanho)

    if tamanho <= 20:
        import pandas as pd
        Tabela = pd.DataFrame(columns=['xi','Fobs','Fa','Fra','zcalc','fesp','|fesp(xi) - fra(xi)|','|fesp(xi) - fra(xi-1)|'])

        tabG = {'0.20':1.07/sqrt(tamanho),'0.15':1.14/sqrt(tamanho),'0.10':1.22/sqrt(tamanho),'0.05':1.36/sqrt(tamanho),'0.025':1.48/sqrt(tamanho),'0.01':1.63/sqrt(tamanho),'0.005':1.73/sqrt(tamanho)}

        Tabela['xi'], Tabela['Fobs'] = np.unique(rol, return_counts=True)
        Tabela['Fa'] = Tabela['Fobs'].cumsum()
        Tabela['Fra'] = Tabela['Fa']/Tabela['Fobs'].sum() # Frequencia relativa acumulada --> Fob.sum = tamanho da rol
        Tabela['zcalc'] = (Tabela['xi'] - Media)/DesvioPadrao
        Tabela['fesp'] = norm.cdf(Tabela['zcalc'])
        Tabela['|fesp(xi) - fra(xi)|'] = abs(Tabela['fesp'] - Tabela['Fra'])
        Tabela['|fesp(xi) - fra(xi-1)|'] = abs(Tabela['fesp'] - Tabela['Fra'].shift(1))
        # Adicionando o primeiro item na coluna |fesp(xi) - fra(xi-1)|
        Tabela.loc[0,'|fesp(xi) - fra(xi-1)|'] = abs(Tabela.loc[0,'fesp'] - 0)

        # Dcalc = Maximo de |fesp(xi) - fra(xi)| e |fesp(xi) - fra(xi-1)|
        Dcalc = max(abs(Tabela['|fesp(xi) - fra(xi)|'].max()), abs(Tabela['|fesp(xi) - fra(xi-1)|'].max()))
        # Usar p-valores (Distribuio Normal inversa e verifica se p-valor >= alpha) ->  Hipotese H1,casso contrario H0

        # Dtab = TabelaG.loc[tamanho-1,alpha]
        Dtab = TabelaG.loc[tamanho-1, str(alpha)]

        if graf:
            Tabela

        # Se Dcalc < Dtab --> Aceita H0
        if Dcalc < Dtab:
            print('\nDcalc < Dtab --> Aceita H0')
            return True

        # Caso contrario --> Rejeita H0
        else:
            print('\nDcalc > Dtab --> Rejeita H0')
            return False

    else:
        # statistica, p_valor = kstest(rol, 'norm', args=(Media,DesvioPadrao))
        statistica, p_valor = shapiro(rol)

        if p_valor >= alpha:
            print('p-valor = ', p_valor)
            print('p-valor >= alpha --> Aceita H0')
            return True

        else:
            print('p-valor = ', p_valor)
            print('p-valor < alpha --> Rejeita H0')
            return False
```

## ShapiroWilk

```python
def ShapiroWilk(rol, alpha, tamanho):
    """
    Funo para realizar o teste de Shapiro-Wilk.

    Parmetros:
        rol (array): Array de dados.
        alpha (float): Nvel de significncia.
        tamanho (int): Tamanho da amostra.

    Retorna:
        bool: True se a hiptese nula  aceita, False caso contrrio.
    """
    print('\nShapiro-Wilk\n')
    print("\nHipotese:\n")
    print("H0 = A amostra segue uma distribuio normal")
    print("H1 = A amostra no segue uma distribuio normal")

    Media = rol.mean()
    Variancia = rol.var()
    DesvioPadrao = rol.std()
    mediana = np.median(rol)
    max = rol.max()
    min = rol.min()
    soma = rol.sum()

    print('Media = ', Media)
    print('Variancia = ', Variancia)
    print('DesvioPadrao = ', DesvioPadrao)
    print("Mediana = ", mediana)
    print("Minimo = ", min)
    print("Maximo = ", max)
    print("Soma = ", soma)
    print("Contagem = ", tamanho)

    HistogramaSimples(rol)

    if tamanho <= 30:
        Tabela = pd.DataFrame(columns=['i','n-(i-1)','ain','xn-(i-1)','xi','ain(xn-(i-1)-xi'])

        Tabela['i'] = np.arange(1,(tamanho+1)/2)
        Tabela['n-(i-1)'] = tamanho - Tabela['i'] + 1
        Tabela['ain'] = TabelaH1.loc[tamanho-1].values[~np.isnan(TabelaH1.loc[tamanho-1].values)] #Soluo Elegante -> usei masks para remover os NaN
        Tabela['xn-(i-1)'] = rol[Tabela['n-(i-1)'].astype(int)-1]
        Tabela['xi'] = rol[Tabela['i'].astype(int)-1]
        Tabela['ain(xn-(i-1)-xi'] = Tabela['ain']*(Tabela['xn-(i-1)']-Tabela['xi'])

        #b = somatorio de ain(xn-(i-1)-xi) (1,n/2)
        #wcalc= b / Somatrio de (xi - x|)

        b = Tabela['ain(xn-(i-1)-xi'].sum()
        Wcalc = (b**2)/((rol - Media)**2).sum()
        Wc = TabelaH2.loc[tamanho-3, str(alpha)]

        print('\nTabela')
        Tabela
        print('\n')

        print('Wcalc = ', Wcalc)
        print('Wc = ', Wc)

        if Wcalc >=  Wc:
            print('\nWcalc >= Wc --> Aceita H0')
            return True
        else:
            print('\nWcalc < Wc --> Rejeita H0')
            return False

    else:
        stat, p = shapiro(rol)

        if p > alpha:
            print('\np > alpha --> Aceita H0')
            return True
        else:
            print('\np < alpha --> Rejeita H0')
            return False
```

## Bartlett


```python
def Bartlett(rol, alpha, graf=True):
    """
    Funo para realizar o teste de Bartlett.

    Parmetros:
        rol (list): Lista de listas de dados.
        alpha (float): Nvel de significncia.
        graf (bool, opcional): Indica se deve exibir grficos e informaes. Padro  True.

    Retorna:
        bool: True se a hiptese nula  aceita, False caso contrrio.
    """
    print("Bartlett\n")
    print('Hipoteses:')
    print('H0 --> As variancias so iguais')
    print('H1 --> As variancias no so iguais')

    # rol = lista de listas

    N = 0
    Ni = []
    Vari = []

    for i in range(len(rol)):
        N += len(rol[i])
        Ni.append(len(rol[i]))
        Vari.append(np.var(rol[i]))

    Sp2 = np.sum((np.array(Ni)-1)*(np.array(Vari)))/(N-len(rol))

    q = (N-len(rol))*np.log(Sp2) - np.sum((np.array(Ni)-1)*np.log(np.array(Vari)))
    c = 1 + (1/(3*(len(rol)-1)))*(np.sum(1/(np.array(Ni)-1))-1/(N-len(rol)))

    Bcalc = q/c

    X2c = chi2.ppf(1-alpha,len(rol)-1)

    if graf:
        HistogramaComparativo(rol)


    if Bcalc > X2c:
        print('\nBcalc > X2c --> Rejeita H0')
        return False

    else:
        print('\nBcalc <= X2c --> Aceita H0')
        return True
```

## Z score

```python
def Zscore(rol, alpha, M):
    """
    Funo para realizar o teste de Z-Score.

    Parmetros:
        rol (array): Array de dados.
        alpha (float): Nvel de significncia.
        M (float): Valor a ser testado.

    Retorna:
        bool: True se a hiptese nula  aceita, False caso contrrio.
    """
    print("Z-Score\n")
    print('Hipoteses:')
    print('H0 --> M = ', M)
    print('H1 --> M != ', M)

    sigmaXb = np.std(rol, ddof=1) / np.sqrt(len(rol))
    zCalc = (np.mean(rol) - M) / sigmaXb

    """
    Os valores crticos da estatstica (zc) so apresentados na Tabela E do apndice.
    Essa tabela fornece os valores crticos de zc tal que P(Zcalc > zc) = 
    (para um teste unilateral  direita).
    """

    zc = t.ppf(1 - alpha / 2, len(rol) - 1)

    HistogramaSimples(rol)

    if abs(zCalc) > zc:
        print('\nZcalc > Zc --> Rejeita H0')
        return False
    else:
        print('\nZcalc <= Zc --> Aceita H0')
        return True
```

## T student Independente

```python
def T_student_Independente(rol1, rol2, alpha):
    """
    Funo para realizar o teste T de Student para comparao de mdias populacionais independentes.

    Parmetros:
        rol1 (array): Primeiro conjunto de dados.
        rol2 (array): Segundo conjunto de dados.
        alpha (float): Nvel de significncia.

    Retorna:
        bool: True se a hiptese nula  aceita, False caso contrrio.
    """
    print("T-Student - Comparao de mdias populacionais Independentes\n")

    print('Hipoteses:')
    print('H0 --> M1 = M2 ')
    print('H1 --> M1 != M2')

    n1, n2 = len(rol1), len(rol2)
    var1, var2 = rol1.var(), rol2.var()
    mi1, mi2 = rol1.mean(), rol2.mean()
    d1, d2 = rol1.std(), rol2.std()
    max1, max2 = rol1.max(), rol2.max()
    min1, min2 = rol1.min(), rol2.min()
    soma1, soma2 = rol1.sum(), rol2.sum()

    print("Mdias:", mi1, mi2)
    print("Varincias:", var1, var2)
    print("Desvios Padres:", d1, d2)
    print("Mximos:", max1, max2)
    print("Mnimos:", min1, min2)
    print("Somas:", soma1, soma2)
    #print("Contagens:",n1,n2)

    HistogramaComparativo([rol1, rol2])

    if Bartlett([rol1, rol2], alpha):

        print("Caso 2 --> Sigma1 == Sigma2\n")

        Sp = sqrt(((n1-1)*var1 + (n2-1)*var2)/n1+n2-2)
        Tcalc = (mi1-mi2)/(Sp*(1/n1 + 1/n2)**1/2)
        Tc = t.ppf(1-alpha/2,n1+n2-2)

    else:
        print("Caso 1 --> Sigma1 != Sigma2\n")

        Tcalc = (mi1 - mi2 )/((var1/n1)+(var2/n2))**2  # Estatistica T
        V = ((var1/n1   +  var2/n2)**2)/   ( ((var1/n1)**2)/(n1-1)  +   ((var2/n2)**2)/(n2-1))  # Graus de liberdade
        Tc = t.ppf(1-alpha/2,V)  # Valor critico

    # Verificar hipoteses

    if Tcalc >= Tc:
        print('\nTalc >= Tc --> Aceita H0')
        return True
    else:
        print('\nTalc < Tc --> Rejeita H0')
        return False
```

## T Studente Emparelhadas

```python
def T_student_rols_Emparelhadas(rol1, rol2, alpha, graf=False):
    """
    Funo para realizar o teste T de Student para comparao de mdias populacionais emparelhadas.

    Parmetros:
        rol1 (array): Primeiro conjunto de dados emparelhados.
        rol2 (array): Segundo conjunto de dados emparelhados.
        alpha (float): Nvel de significncia.
        graf (bool, optional): Indica se deve exibir grficos. Padro  False.

    Retorna:
        bool: True se a hiptese nula  aceita, False caso contrrio.
    """
    print("T-Student - Comparao de mdias populacionais Emparelhadas\n")

    print('Hipoteses:')
    print('H0 --> Md = 0 ')
    print('H1 --> Md != 0')

    # Verifica se as amostras seguem uma distribuio normal
    if KolmogorovSmirnov(rol1, alpha, len(rol1), graf=False) and KolmogorovSmirnov(rol2, alpha, len(rol2), graf=False):

        # Verifica se as amostras tm varincias iguais
        if Bartlett([rol1, rol2], alpha, graf):

            tamanho = (len(rol1) + len(rol2)) / 2
            rolD = rol1 - rol2  # Diferena entre os pares emparelhados
            Dmean = rolD.mean()  # Mdia das diferenas
            Sd = ((rolD - Dmean)**2 / (tamanho - 1))**0.5  # Desvio padro das diferenas
            Tcalc = Dmean / (Sd / (tamanho**0.5))  # Estatstica T

            Tc = t.ppf(1 - alpha/2, tamanho - 1)  # Valor crtico

            print('Tc = ', Tc)
            print('Tcalc = ', Tcalc)

            if graf:
                HistogramaComparativo([rol1, rol2])

            if abs(Tcalc) > Tc:
                print('\n|Tcalc| > Tc --> Rejeita H0')
                return False
            else:
                print('\n|Tcalc| <= Tc --> Aceita H0')
                return True

        else:
            print("Amostras no possuem varincias iguais")

    else:
        print("Ambas as amostras no provm de uma mesma distribuio (Normal)")
```
